## MPPNP core-H Burn example

Here we analyse the post-processed core-He burning phase of a stellar model of a 5$M_\odot$ at $Z=0.02$ from Set1.

```python
%pylab nbagg
import nugridse as mp
```

```python
sefiles=mp.se('../H5_out')
```

### Evolution of central abundances

H destruction and He production via the CNO cycle

```python
plt.figure(1)
cyclerange=range(500,1001,20)
h1=sefiles.get(cyclerange,'H-1')
he4=sefiles.get(cyclerange,'He-4')
c12=sefiles.get(cyclerange,'C-12')
n14=sefiles.get(cyclerange,'N-14')

h1_c=[]
he4_c=[]
c12_c=[]
n14_c=[]
for k in range(len(cyclerange)):
        h1_c.append(h1[k][0])
        he4_c.append(he4[k][0])
        c12_c.append(c12[k][0])
        n14_c.append(n14[k][0])

markevery=50
plt.plot(cyclerange,h1_c,label='H-1',marker='o',markevery=markevery)
plt.plot(cyclerange,he4_c,label='He-4',marker='s',markevery=markevery)
plt.plot(cyclerange,c12_c,label='C-12',marker='^',markevery=markevery)
plt.plot(cyclerange,n14_c,label='N-14',marker='>',markevery=markevery)
plt.yscale('log')
plt.xlabel('Cycle number')
plt.ylabel('log-scaled X$_i$')
plt.legend(loc=2)
plt.title('Abundance evolution in center')
```

### Abundance profiles of last output

H is depleted in the central region.

```python
plt.figure(2)
cycle=1000
h1=sefiles.get(cycle,'H-1')
he4=sefiles.get(cycle,'He-4')
c12=sefiles.get(cycle,'C-12')
n14=sefiles.get(cycle,'N-14')
m=sefiles.get(cycle,'mass')
markevery=50
plt.plot(m,h1,label='H-1',marker='o',markevery=markevery)
plt.plot(m,he4,label='He-4',marker='s',markevery=markevery)
plt.plot(m,c12,label='C-12',marker='^',markevery=markevery)
plt.plot(m,n14,label='N-14',marker='>',markevery=markevery)
plt.yscale('log')
plt.xlabel('M/Msun')
plt.ylabel('log-scaled X$_i$')
plt.legend(loc=1)
plt.title('Abundance profile of cycle 1000')

```
