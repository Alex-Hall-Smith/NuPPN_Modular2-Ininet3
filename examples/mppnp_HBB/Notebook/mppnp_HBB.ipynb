## MPPNP TP-AGB phase with efficient HBB

Intermediate-mass stars, in particular at low Z, undergo hot-bottom-burning (HBB) in their envelopes. Temperatures
at the bottom of the envelopes reach high enough to active the CNO cycle. Here we analyze the hot envelope of a   6$M_\odot$ at $Z=0.01$ from Set1.

```python
%pylab nbagg
import nugridse as mp
```

```python
sefiles=mp.se('../H5_out')
```

### Initial TP-AGB profile

```python
plt.close(1)
cycle=4800
h1=np.log10(sefiles.get(cycle,'H-1'))
he4=np.log10(sefiles.get(cycle,'He-4'))
c12=np.log10(sefiles.get(cycle,'C-12'))
o16=np.log10(sefiles.get(cycle,'O-16'))
m=sefiles.get(cycle,'mass')
```

```python
plt.figure(1)
markevery=50
plt.plot(m,h1,label='H-1',marker='x',markevery=markevery,linestyle='-.')
plt.plot(m,he4,label='He-4',marker='o',markevery=markevery,linestyle='--')
plt.plot(m,c12,label='C-12',marker='D',markevery=markevery,linestyle=':')
plt.plot(m,o16,label='O-16',marker='^',markevery=markevery,linestyle='-')
plt.legend(loc=1)
#plt.yscale('log')
plt.xlabel("M/Msun")
plt.ylabel('log X$_i$')
plt.ylim(1e-1,1)
```

### Surface evolution

Effective destruction of C12 and production of C13 via the CN cycle leads to the following decreasing surface ratios.

```python
sefiles=mp.se('../H5_surf')
```

```python
c12=np.array(sefiles.get('C-12'))
n13=sefiles.get('N-13')
n14=np.array(sefiles.get('N-14'))
c13=np.array(sefiles.get('C-13'))
o16=np.array(sefiles.get('O-16'))
cycs=sefiles.se.cycles
```
